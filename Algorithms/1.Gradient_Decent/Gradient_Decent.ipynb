```python
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
# Generate data using dummy function
X=np.arange(10);
Y=(X-5)**2+4;
print(X,Y);
```

```python
# visulaize

plt.style.use("seaborn");
plt.ylabel("Y");
plt.xlabel("X");
plt.plot(X,Y);
plt.show();
```

```python
# Gradient Descent Algorithm
x=-9;
lr=0.1;
plt.plot(X,Y);
for i in range(30):
    # differentiate the function wrt x f(x)=(x-5)^2
    grad=2*(x-5);
    x=x-lr*grad;
    y=(x-5)**2+4;
    plt.scatter(x,y);
    print(x,y);
```
