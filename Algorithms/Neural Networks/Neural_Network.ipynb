```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons,make_circles,make_classification
```

```python
def load_dataset(dataset):
    if dataset=='moons':
        X,Y=make_moons(n_samples=10000,noise=0.2,random_state=1)
    elif dataset=='circles':
        X,Y=make_circles(n_samples=10000,noise=0.1,random_state=1,factor=0.5)
    elif dataset=='classification':
        X,Y=make_classification(n_samples=10000,n_features=2,n_redundant=0,random_state=1)
    return X,Y
```

```python
dataset='classification'
X,Y=load_dataset(dataset)
print(X.shape,Y.shape)
plt.scatter(X[:,0],X[:,1],c=Y)
```

### Model Building

```python
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
# play with the number of neurons in the first layer and see the difference in the decision boundary
# input_shape is only required for the first layer
# activation function is relu


model.add(Dense(10,activation='relu',input_shape=(2,))) 

model.add(Dense(5,activation='relu'))

# convert the number to probability using sigmoid
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',metrics='accuracy')
model.summary()
```

```python
model.fit(X,Y,batch_size=64,epochs=50,validation_split=0.1)
```
